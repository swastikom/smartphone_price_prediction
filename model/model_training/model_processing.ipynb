```python
import pandas as pd
import numpy as np

df = pd.read_csv('../clean_dataset/clean.csv')
df = df[['os','Number of Sim','Processor Name','Processor No. of Cores','Ram Size(GB)','Rom Size(GB)','Battery Capacity','Display Size(inches)','Rear Camera 1','Front Camera','External Card Support','Brand','price']]
df
```

```python
from sklearn.preprocessing import LabelEncoder

os_encoder = LabelEncoder()
brand_encoder = LabelEncoder()
processor_encoder = LabelEncoder()

os_encoder.fit(df['os'])
brand_encoder.fit(df['Brand'])
processor_encoder.fit(df['Processor Name'])

df['os'] = os_encoder.transform(df['os'])
df['Brand'] = brand_encoder.transform(df['Brand'])
df['Processor Name'] = processor_encoder.transform(df['Processor Name']) 

df
```

```python
X = df.drop(['Brand', 'price'], axis=1)
y_price = df['price']
X
```

```python

from sklearn.model_selection import train_test_split
```

```python


# Instantiate and train a GradientBoostingRegressor for price prediction
while True:
    from sklearn.ensemble import GradientBoostingRegressor
    X_price_train, X_price_test, y_price_train, y_price_test = train_test_split(X, y_price, test_size=0.2)
    price_gbr = GradientBoostingRegressor(n_estimators=200, random_state=50)
    price_gbr.fit(X_price_train, y_price_train)
    score = price_gbr.score(X_price_test, y_price_test)
    print('Bad score : ')
    print(score)
    if 0.9 < score:
        break
    
print(score)


# Make predictions

```

```python
import pickle
import numpy as np

with open('../models/price_model.pkl', 'wb') as file:
    pickle.dump((price_gbr, os_encoder, processor_encoder), file)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate predicted prices using the trained Gradient Boosting Regressor
predicted_prices_train = price_gbr.predict(X_price_train)

# Sort the actual and predicted prices based on actual prices
sorted_indices = np.argsort(y_price_train)
sorted_actual_prices = y_price_train.iloc[sorted_indices]
sorted_predicted_prices = predicted_prices_train[sorted_indices]

# Plot the regression line
plt.figure(figsize=(10, 6))
plt.scatter(sorted_actual_prices, sorted_predicted_prices,
            color='red', label='Predicted',marker='+')
plt.plot(sorted_actual_prices, sorted_actual_prices,
         color='blue', linestyle=':', label='Ideal')

plt.xlabel('Actual Prices')
plt.ylabel('Predicted Prices')
plt.title('Regression Line: Actual vs. Predicted Prices (Training Samples)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
df2 = df[['Processor Name','Battery Capacity','Front Camera', 'Display Size(inches)','Number of Sim', 'Brand']]
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

X = df2.drop('Brand', axis=1)
y = df2['Brand']

while True:
    # Split the dataset into training and testing sets
    X_train, X_test, y_train, y_test = train_test_split(
        X, y, test_size=0.1, random_state=45)

    # Initialize the Random Forest classifier
    clf = RandomForestClassifier()

    # Train the classifier on the training data
    clf.fit(X_train, y_train)

    # Calculate the accuracy of the model on the test data
    accuracy = clf.score(X_test, y_test)
    y_pred = clf.predict(X_test)
    print('Current accuracy:', accuracy)

    # Check if the accuracy is greater than 0.9
    if accuracy > 0.7:
        break


```

```python
import pickle


# Export the model and label encoder using pickle
with open('../models/brand_model.pkl', 'wb') as file:
    pickle.dump((clf, brand_encoder, processor_encoder), file)
```

```python
plt.figure(figsize=(10, 6))

plt.hist(y_test, alpha=0.5, label='Test Values', bins=20, color='blue')
plt.hist(y_pred, alpha=0.5, label='Predicted Values', bins=20, color='green')

plt.xlabel('Brand')
plt.ylabel('Frequency')
plt.title('Histogram of Test Values and Predicted Values')
plt.legend()
plt.show()
```
